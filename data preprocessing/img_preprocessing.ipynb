```python
img_name = '/img.jpg'

face_cascade = cv2.CascadeClassifier('C:/Anaconda3/Lib/site-packages/cv2/data/haarcascade_frontalface_default.xml')
img = cv2.imread('{}.jpg'.format(img_name))
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray, 2, 5)

for (x,y,w,h) in faces:
    src = cv2.imread("{}.jpg".format(img_name), cv2.IMREAD_COLOR)
    dst = src.copy() 
    dst = src[y-30:(y+h)+30,x:(x+w)]
#     cv2.imshow("src", src)
#     cv2.imshow("dst", dst)
    cv2.waitKey(0)
    cv2.imwrite('images/face.jpg', dst)
#     cv2.destroyAllWindows()
```

```python
# bilateral 필터
img = cv2.imread('img.jpg')

blur = cv2.bilateralFilter(img,5,90,90)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
cv2.imwrite('images/{}.jpg'.format('face_blurred'), blur)
```

```python
from PIL import Image, ImageDraw
import face_recognition
```

```python
# Load the jpg file into a numpy array
image = cv2.imread("jo.jpg")

# Find all facial features in all the faces in the image
face_landmarks_list = face_recognition.face_landmarks(image)

pil_image = Image.fromarray(image)
for face_landmarks in face_landmarks_list:
    d = ImageDraw.Draw(pil_image, 'RGBA')

    # Make the eyebrows into a nightmare
    d.polygon(face_landmarks['left_eyebrow'], fill=(68, 54, 39, 128))
    d.polygon(face_landmarks['right_eyebrow'], fill=(68, 54, 39, 128))
    d.line(face_landmarks['left_eyebrow'], fill=(68, 54, 39, 150), width=5)
    d.line(face_landmarks['right_eyebrow'], fill=(68, 54, 39, 150), width=5)

    # Gloss the lips
    d.polygon(face_landmarks['top_lip'], fill=(150, 0, 0, 128))
    d.polygon(face_landmarks['bottom_lip'], fill=(150, 0, 0, 128))
    d.line(face_landmarks['top_lip'], fill=(150, 0, 0, 64), width=8)
    d.line(face_landmarks['bottom_lip'], fill=(150, 0, 0, 64), width=8)

    # Sparkle the eyes
    d.polygon(face_landmarks['left_eye'], fill=(255, 255, 255, 30))
    d.polygon(face_landmarks['right_eye'], fill=(255, 255, 255, 30))

    # Apply some eyeliner
    d.line(face_landmarks['left_eye'] + [face_landmarks['left_eye'][0]], fill=(0, 0, 0, 110), width=6)
    d.line(face_landmarks['right_eye'] + [face_landmarks['right_eye'][0]], fill=(0, 0, 0, 110), width=6)

    pil_image.show()
```
