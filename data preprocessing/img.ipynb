```
# file_path = tf.keras.utils.get_file('수지.jpg', '/content/drive/My Drive/project3/수지.jpg')
image = cv2.imread('/content/drive/My Drive/project3/수지.jpg')
#image = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("/content/drive/My Drive/project3/shape_predictor_68_face_landmarks.dat")

faces = detector(image)
for face in faces:
    x1, y1, x2, y2 = face.left(), face.top(), face.right(), face.bottom()
    #cv2.rectangle(image, (x1, y1), (x2, y2), (0, 0, 255), thickness=2)

    landmarks = predictor(image, face)
    for n in range(0, 68):
        x = landmarks.part(n).x
        y = landmarks.part(n).y
        cv2.circle(image, (x, y), 4, (255, 0, 0), -1)
        cv2.putText(image, str(n), (x, y), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,255,255), 1, cv2.LINE_AA)


# cv2.imshow('image', image)
# cv2.waitKey(0) 
# cv2.destroyWindow('image')

plt.imshow(image)
plt.show()
```
